````
# OpenAlex series for 1960–1995: Output, yNCI (year-normalized), %Top10
# Runs entirely in Colab. It saves 2 CSVs to /content/.

import requests, time, pandas as pd, numpy as np

MAILTO = "ehsan.roohi.fum@gmail.com"     # <-- ایمیل خودت
BASE   = "https://api.openalex.org"
SLEEP  = 0.25
START, END = 1960, 1995
DOC_TYPES = ["journal-article", "review-article"]

# کشورها (اگر لازم داری کم/زیاد کن)
COUNTRIES = {
    "IR":"Iran","TR":"Turkey","EG":"Egypt","PK":"Pakistan","SA":"Saudi Arabia",
    "AE":"United Arab Emirates","MY":"Malaysia","ID":"Indonesia","TH":"Thailand","VN":"Vietnam",
    "KR":"South Korea","TW":"Taiwan","SG":"Singapore","IL":"Israel","GR":"Greece",
    "NL":"Netherlands","CN":"China"
}

def fetch_all(url, params):
    p = params.copy(); p["per-page"]=200; p["cursor"]="*"; p["mailto"]=MAILTO
    out=[]
    while True:
        r=requests.get(url,params=p,timeout=60); r.raise_for_status()
        j=r.json(); res=j.get("results",[]); out+=res
        cur=j.get("meta",{}).get("next_cursor")
        if not cur or not res: break
        p["cursor"]=cur; time.sleep(SLEEP)
    return out

def c5(counts_by_year, y):
    if not counts_by_year: return 0
    hi=y+4; s=0
    for row in counts_by_year:
        yy=row.get("year"); cc=row.get("cited_by_count",0) or 0
        if yy is not None and y<=yy<=hi: s+=cc
    return s

# --- 1) baseline جهانی سالانه (بدون رشته)
baseline=[]
for y in range(START, END+1):
    vals=[]
    for dt in DOC_TYPES:
        W=fetch_all(f"{BASE}/works", {
            "filter": f"from_publication_date:{y}-01-01,to_publication_date:{y}-12-31,type:{dt}",
            "select": "publication_year,counts_by_year"
        })
        for w in W:
            py=w.get("publication_year",y) or y
            vals.append(c5(w.get("counts_by_year"), py))
    vals=np.array(vals,dtype="float64")
    baseline.append({
        "year":y, "n_world":int(vals.size),
        "median5": float(np.median(vals)) if vals.size else 0.0,
        "p90_5":   float(np.percentile(vals,90)) if vals.size else 0.0
    })
baseline=pd.DataFrame(baseline)
baseline.to_csv("/content/baseline_world_year_1960_1995.csv", index=False)

# --- 2) کشور به تفکیک سال: خروجی و کیفیتِ سال-نرمال
rows=[]
for cc,name in COUNTRIES.items():
    for y in range(START, END+1):
        vals=[]
        for dt in DOC_TYPES:
            W=fetch_all(f"{BASE}/works", {
                "filter": f"from_publication_date:{y}-01-01,to_publication_date:{y}-12-31,"
                          f"authorships.institutions.country_code:{cc},type:{dt}",
                "select":"publication_year,counts_by_year"
            })
            for w in W:
                py=w.get("publication_year",y) or y
                vals.append(c5(w.get("counts_by_year"), py))
        rows.append({"country_code":cc,"country":name,"year":y,"c5_list":vals,"n_output":len(vals)})

df=pd.DataFrame(rows).merge(baseline, on="year", how="left")

def ynci_mean(lst, med):
    if med<=0 or not lst: return np.nan
    arr=np.array(lst,dtype="float64")
    return float(np.nanmean(arr/med))

def pct_top10(lst, p90):
    if p90<=0 or not lst: return np.nan
    arr=np.array(lst,dtype="float64")
    return float((arr>=p90).mean()*100.0)

df["yNCI"]      = [ynci_mean(lst, med) for lst,med in zip(df["c5_list"], df["median5"])]
df["pct_top10"] = [pct_top10(lst, p90) for lst,p90 in zip(df["c5_list"], df["p90_5"])]
df.drop(columns=["c5_list"], inplace=True)
df.to_csv("/content/country_year_1960_1995_output_yNCI_top10.csv", index=False)

print("Saved:")
print(" - /content/baseline_world_year_1960_1995.csv")
print(" - /content/country_year_1960_1995_output_yNCI_top10.csv")
````

````
import requests, time, pandas as pd, numpy as np

MAILTO = "ehsan.roohi.fum@gmail.com"
BASE   = "https://api.openalex.org"
SLEEP  = 0.5
START, END = 1960, 1995

# اگر خواستی کشور کم/زیاد کن
COUNTRIES = {
    "IR":"Iran","TR":"Turkey","EG":"Egypt","PK":"Pakistan","SA":"Saudi Arabia",
    "AE":"United Arab Emirates","MY":"Malaysia","ID":"Indonesia","TH":"Thailand","VN":"Vietnam",
    "KR":"South Korea","TW":"Taiwan","SG":"Singapore","IL":"Israel","GR":"Greece",
    "NL":"Netherlands","CN":"China"
}

def fetch_all(url, params, retries=3):
    p = params.copy()
    p["per-page"] = 200
    p["cursor"]   = "*"
    p["mailto"]   = MAILTO
    out = []
    while True:
        try:
            r = requests.get(url, params=p, timeout=60)
            if r.status_code in (403, 429, 502, 503) and retries>0:
                time.sleep(2.0); retries -= 1; continue
            r.raise_for_status()
        except Exception as e:
            # در صورت خطا کمی صبر کن و دوباره
            if retries>0:
                time.sleep(2.0); retries -= 1; continue
            else:
                raise e
        j = r.json()
        res = j.get("results", [])
        out.extend(res)
        cur = j.get("meta", {}).get("next_cursor")
        if not cur or not res:
            break
        p["cursor"] = cur
        time.sleep(SLEEP)
    return out

def c5(counts_by_year, pub_year, window=5):
    if not counts_by_year:
        return 0
    hi = pub_year + window - 1
    s = 0
    for row in counts_by_year:
        y = row.get("year")
        c = row.get("cited_by_count", 0) or 0
        if y is not None and pub_year <= y <= hi:
            s += c
    return s

# ---------- 1) Baseline جهانی سالانه (بدون رشته) ----------
baseline_rows = []
for y in range(START, END+1):
    vals = []
    # فقط journal-article + حذف paratext
    params = {
        "filter": f"from_publication_year:{y},to_publication_year:{y},"
                  f"type:journal-article,is_paratext:false",
        "select": "publication_year,counts_by_year,cited_by_count"
    }
    works = fetch_all(f"{BASE}/works", params)
    for w in works:
        py = w.get("publication_year", y) or y
        vals.append(c5(w.get("counts_by_year"), py))
    vals = np.array(vals, dtype="float64")
    baseline_rows.append({
        "year": y,
        "n_world": int(vals.size),
        "median5": float(np.median(vals)) if vals.size else 0.0,
        "p90_5":   float(np.percentile(vals, 90)) if vals.size else 0.0
    })
baseline = pd.DataFrame(baseline_rows)
baseline.to_csv("/content/baseline_world_year_1960_1995.csv", index=False)
print("Baseline sample:\n", baseline.head())

# ---------- 2) Country × Year ----------
rows = []
for cc, name in COUNTRIES.items():
    for y in range(START, END+1):
        vals = []
        params = {
            "filter": f"from_publication_year:{y},to_publication_year:{y},"
                      f"authorships.institutions.country_code:{cc},"
                      f"type:journal-article,is_paratext:false",
            "select": "publication_year,counts_by_year"
        }
        wk = fetch_all(f"{BASE}/works", params)
        for w in wk:
            py = w.get("publication_year", y) or y
            vals.append(c5(w.get("counts_by_year"), py))
        rows.append({"country_code": cc, "country": name, "year": y,
                     "c5_list": vals, "n_output": len(vals)})

df = pd.DataFrame(rows).merge(baseline, on="year", how="left")

def ynci_mean(lst, med):
    if med <= 0 or not lst:
        return np.nan
    arr = np.array(lst, dtype="float64")
    return float(np.nanmean(arr / med))

def pct_top10(lst, p90):
    if p90 <= 0 or not lst:
        return np.nan
    arr = np.array(lst, dtype="float64")
    return float((arr >= p90).mean() * 100.0)

df["yNCI"]      = [ynci_mean(lst, med) for lst, med in zip(df["c5_list"], df["median5"])]
df["pct_top10"] = [pct_top10(lst, p90) for lst, p90 in zip(df["c5_list"], df["p90_5"])]
df.drop(columns=["c5_list"], inplace=True)
df.to_csv("/content/country_year_1960_1995_output_yNCI_top10.csv", index=False)

print("Saved 2 CSVs in /content/. Nonzero checks:")
print(df.groupby("country")["n_output"].sum().sort_values(ascending=False).head(8))
````

````
from pyalex import Works
import numpy as np, pandas as pd
from tqdm import tqdm

START, END = 1960, 1995

def c5(counts_by_year, y):
    if not counts_by_year: return 0
    hi = y + 4
    s = 0
    for row in counts_by_year:
        yy = row.get("year"); cc = row.get("cited_by_count", 0) or 0
        if yy is not None and y <= yy <= hi: s += cc
    return s

baseline_rows = []
for y in tqdm(range(START, END+1), desc="Baseline"):
    q = (Works()
         .filter(publication_year=y, type="journal-article", is_paratext=False)
         .select(["publication_year","counts_by_year"]))
    vals = []
    for page in q.paginate(per_page=200):   # ← این مهم است
        for w in page:
            py = w.get("publication_year", y) or y
            vals.append(c5(w.get("counts_by_year"), py))

    arr = np.array(vals, dtype="float64")
    baseline_rows.append({
        "year": y,
        "n_world": int(arr.size),
        "median5": float(np.median(arr)) if arr.size else 0.0,
        "p90_5":   float(np.percentile(arr, 90)) if arr.size else 0.0
    })

baseline = pd.DataFrame(baseline_rows)
display(baseline.head())
baseline.to_csv("/content/baseline_world_year_1960_1995.csv", index=False)
````

````
# Crossref 1960–1995: baseline جهانی + کشور×سال (approx yNCI*, %Top10*)
import requests, time, math, pandas as pd, numpy as np
from tqdm import tqdm

EMAIL = "ehsan.roohi.fum@gmail.com"                  # ایمیل خودت
UA    = f"Mozilla/5.0; mailto:{EMAIL}"               # User-Agent مودبانه
BASE  = "https://api.crossref.org/works"
SLEEP = 0.4                                          # سرعت امن
START, END = 1960, 1995

# 8 کشور مقاله
COUNTRIES = {
    "Iran": "Iran",
    "China": "China",
    "South Korea": "South Korea",
    "Taiwan": "Taiwan",
    "Singapore": "Singapore",
    "Israel": "Israel",
    "Greece": "Greece",
    "Netherlands": "Netherlands",
}

session = requests.Session()
session.headers.update({"User-Agent": UA, "Accept": "application/json"})

def cr_cursor_pages(params, max_hits=None):
    """Crossref deep paging with cursor=*; yields item dicts."""
    p = params.copy()
    p["rows"] = 1000
    p["cursor"] = "*"
    seen = 0
    while True:
        r = session.get(BASE, params=p, timeout=90)
        if r.status_code != 200:
            # چاپ خطا برای عیب‌یابی
            print("HTTP", r.status_code, r.text[:200], "\nURL:", r.url)
            break
        j = r.json()
        items = j.get("message", {}).get("items", [])
        if not items:
            break
        for it in items:
            yield it
            seen += 1
            if max_hits and seen >= max_hits:
                return
        p["cursor"] = j.get("message", {}).get("next-cursor")
        if not p["cursor"]:
            break
        time.sleep(SLEEP)

def issued_year(rec):
    # Crossref issued: {"date-parts": [[YYYY, M, D]]}
    try:
        return int(rec.get("issued", {}).get("date-parts", [[None]])[0][0])
    except Exception:
        return None

def citations(rec):
    # کل استنادات تا امروز
    return int(rec.get("is-referenced-by-count", 0) or 0)

def filter_common(year, country_query=None):
    f = {
        "from-pub-date": f"{year}-01-01",
        "until-pub-date": f"{year}-12-31",
        "type": "journal-article",
        "has-affiliation": "true",   # برای کشور لازم است
    }
    # انتخاب محدود فیلدها برای سرعت
    sel = "DOI,issued,is-referenced-by-count,author,container-title"
    params = {"filter": ",".join([f"{k}:{v}" for k,v in f.items()]),
              "select": sel, "mailto": EMAIL}
    # اگر کشور خواستیم، از query.affiliation استفاده می‌کنیم
    if country_query:
        params["query.affiliation"] = country_query
    return params
````

````
baseline_rows = []
for y in tqdm(range(START, END+1), desc="Baseline (Crossref)"):
    params = filter_common(y)  # بدون country
    vals = []
    for rec in cr_cursor_pages(params):
        py = issued_year(rec)
        if py != y:  # ایمنی
            continue
        vals.append(citations(rec))
    arr = np.array(vals, dtype="float64")
    baseline_rows.append({
        "year": y,
        "n_world": int(arr.size),
        "median_cit": float(np.median(arr)) if arr.size else 0.0,
        "p90_cit": float(np.percentile(arr, 90)) if arr.size else 0.0
    })

baseline = pd.DataFrame(baseline_rows)
display(baseline.head())
baseline.to_csv("/content/baseline_world_year_1960_1995_crossref.csv", index=False)
````

````
bl = pd.read_csv("/content/baseline_world_year_1960_1995_crossref.csv")

rows = []
for cname in COUNTRIES.keys():
    for y in tqdm(range(START, END+1), desc=cname):
        params = filter_common(y, country_query=cname)
        cits = []
        for rec in cr_cursor_pages(params):
            py = issued_year(rec)
            if py != y:
                continue
            cits.append(citations(rec))
        n = len(cits)
        med = float(bl.loc[bl.year==y, "median_cit"].values[0]) if (bl.year==y).any() else 0.0
        p90 = float(bl.loc[bl.year==y, "p90_cit"].values[0]) if (bl.year==y).any() else 0.0
        yNCI_star = float(np.mean(np.array(cits, dtype="float64")/med)) if (n>0 and med>0) else np.nan
        pct10_star= float((np.array(cits, dtype="float64")>=p90).mean()*100.0) if (n>0 and p90>0) else np.nan
        rows.append({"country": cname, "year": y,
                     "n_output": n, "yNCI_star": yNCI_star, "pct_top10_star": pct10_star})

df = pd.DataFrame(rows)
display(df.head())
df.to_csv("/content/country_year_1960_1995_output_yNCIstar_top10star_crossref.csv", index=False)
````
